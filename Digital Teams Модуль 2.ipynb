**Основные задачи NLP**

**Как преобразовать слова в вектора?**

- Один из вариантов это one-hot encoding но недоастаток в
длина векторов и их ортогональности
- Word2Vec - семантические вектора (например, king - man + woman = queen), использует всего 300 компонент на вектор (гиперпараметр). Интерпретируемости у таких компонент нет.

Для обучения **Word2Vec** используются классические алгоритмы ML (например, градиентный спуск) плюс специфичный алгоритм вычисления весов (skip gram или continious BOW, т.е. предсказание контекста по слову или слова по контексту). Неизвестные слова заменяются на пустой вектор.

**Языковое моделирование (Language Modeling)**

Count-based Models - статистические модели (аля Т9)
Naural Language Models - нейронные модели

В качестве функции потерь чатсо применяют функцию перекрестной жнтропии (Cross-Entropy).

## Лекция 9 - Retrieval-Augmented Generation (RAG)

Основная идея в том, чтобы обогатить запрос к БЯМ дополнительной информацией из базы знаний.
Это помогает устранять галлюцинации.

Улучшает точность и релеватность, может приводить ссылки на источники фактов.
Дает возмодность отвечать "я не знаю" вместо того чтобы выдумывать ответ.

Используется векторная база данных.

**Два способа кодировать документы и запросы:**

* **Bi-Encoder**
Запросы кодируются альтернативной моделью относстельно документов так, чтобы ответы были близки к запросу по косинусной близости.
  
 Минус: по качеству уступает Cross-Encoder'у.

* **Cross-Encoder**
Кодируем моделью сразу запрос и релеватный ответ.

 Минус: нет возмонжости кеширования.

**Frozen RAG** - в этом лучае ретривер и БЯМ не дообучаются.

Альтернатива: обучаем ретривер для генератора.

Третий способ - обучаем, как ретривер, так и БЯМ.

Аткуальный способ: **Advanced Frozen RAG**

**Модулярный RAG**

**Оценка RAG**
- реалистичность
- разнообразие
- глубина анализа
- надежность
- долговечность

Метрики ранжирования, метрики основанные на именованных сущностях, генеративные метрики.

**Ragas** - фреймворк для оценки RAG
- достоверность
- релевантность
- точность контекста (context precision)
- полнота контекста (context recall)

**Обучение с подкреплением**

Это обучение интеллектуального агента хорошей последовательности принятия решения в среде с **неполной информацией**.

**Марковский процесс принятия решения.**

Фактически все сводится к разработке функции, которая для множества состояний даст нужное действие.

**Набледение** - это часть состояния.
Восстанавливаем состояние из наблюдения с помощью, например, рекурентной нейронной сети.

**Функция полезности** - это фактически предсказание того, что получим в будущем относительно текущего состояния.

**Уравнение Беллмана** используется для оптимизации функции полезноести.

Ключевое отличие обучения с подкреплением в том, что целевое значение постоянно меняется. Это и ключевая проблема обучения с подкреплением.

**Алгоритм актор-критик**

Критик отключается после настройки актора.

**Исользование опыта** (отложенный off-policy или актуальный самый свежий on-policy - сходится надежней)

Нужно оставлять зазор на продолжение обучения, т.к. агент никогла не знает полного состояния окружающей среды. Нужно делать неоптимальные действия (что-то случайное).

ML находится на стыке обучения с учителем обучения, без учителя и обучения с подкреплением.

RL это мягкая оптимизация (поверхность функции потерь неровная и постоянно изменяется).
