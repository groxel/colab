# Цель статьи

Разберемся на простейших примерах, когда уместно использовать логарифмическое преобразование целевой переменной в задаче линейной регрессии, а когда можно этого не делать.

# Постановка задачи

Дан набор наблюдений вида $\{(x_1, y_1), (x_2,y_2), \dots,(x_n,y_n)\}$, где $x,y \in \mathbb{R}$ (т.е. x,y в нашем случае просто числа, рассматриваем одномерную регрессию с одним параметром). По этому набору мы хотим построить функцию $a(x)$, которая по заданному $x$ вычисляет значение $y$. Предпологается, что $y$(целевая переменная) связанна с переменной $x$ таким способом:

$$y =b+w*x+\epsilon$$, 

где $\epsilon$ - случайная величина, имеющая нормальное распределение с параметрами $N(\mu, \sigma)$, $w$ - неизвестный коэффициент, который надо оценить из данных. Для упрощения будем считать, что $b=0, \mu=0$, поэтому $y =w*x+\epsilon$. Так же считаем, что все наблюдения в выборке независимы.

# Решение задачи

Известно, что если b - это число, а $\epsilon$ - случайная величина с нормальным распределением $N(\mu, \sigma)$, то сумма $b+\epsilon$ есть так же случайная величина с распределением $N(\mu+b, \sigma)$, т.е. прибавляя константу к случайной величине мы сдвигаем это распределение.
Т.к. $y =w*x+\epsilon$, то $y$ - случайная величина с нормальным распределением $N(w*x, \sigma)$, а плотность распределения будет иметь вид: 

$$f(y) = \frac{1}{\sigma\sqrt{2\pi}}exp\left(-\frac{(y-w*x)^2}{2\sigma^2}\right)$$

Поскольку в нашей выборке есть $\{y_1, y_2, \dots,y_n\}$, то мы хотели бы так подобрать параметр w, чтобы вероятность увидеть именно такую выборку как у нас была бы максимальной( $\href{https://en.wikipedia.org/wiki/Maximum_likelihood_estimation}{MLE}$). Поскольку y - непрерывная случайная величина, то максимизировать будем не вероятность, а плотность. Совместная плотность нашей выборки есть произведение плотностей каждой $y_i$

$$f(y_1)*f(y_2)*\dots*f(y_n) = \frac{1}{\sigma\sqrt{2\pi}}exp(-\frac{(y_1-w*x_1)^2}{2\sigma^2}) * \frac{1}{\sigma\sqrt{2\pi}}exp(-\frac{(y_2-w*x_2)^2}{2\sigma^2}) * \dots$$ 

Обозначим это выражение как $L(w)$, которое мы будем рассматривать относительно параметра w. $L(w)$ - называется функцией правдоподобия и ее мы будем максимизировать по параметру w. Теперь мы перемножаем все n слагаемых и упрощаем наше выражение, получив следующее:

$$L(w) = \left(\frac{1}{\sigma\sqrt{2\pi}}\right)^n exp\left(-\frac{1}{2\sigma^2}\sum_{i=1}^{n}(y_i-w*x_i)^2\right)$$

Мы можем убрать константу $ \left(\frac{1}{\sigma\sqrt{2\pi}}\right)^n$, поскольку это никак не повлияет на максимизацию, а так же взять логарифм натуральный от $L(w)$, поскольку логарифм это монотонная функция и она так же не повлияет на результаты, но выражение станет чуть приятнее:

$$\log{L(w)} = -\frac{1}{2\sigma^2}\sum_{i=1}^{n}(y_i-w*x_i)^2$$

Множитель перед суммой так же можем выбросить - не влияет на поиск оптимальных параметров и получим такую задачу - найти такое значение параметра w, которое максимизирует нашу функцию, или кратко:

$$ w = arg\max_{w \in \mathbb{R}} \left( -\sum_{i=1}^{n}(y_i-w*x_i)^2\right) $$

Далее можем пытаться решить задачу аналитически(найти производную и приравнять к нулю), использовать градиентный спуск или записать функцию правдоподобия в виде функции на python и передать оптимизатору из scipy. Я выбрал последнее.

В контексте нашей темы мы можем сделать следующие **полезные выводы**:

1. В нашей модели каждое значение целевой переменной $y_i$ имеет нормальное распределение. Но ни откуда не следует, что построив гистограмму для $\{y_1, y_2, \dots,y_n\}$ она обязательно окажется нормальной(мы увидим это дальше в примерах).

2. В модели не подразумевается, что все $x_i$ будут различны, а на практике часто так и происходит, что одному $x_j$, соответствует сразу несколько ${y_{j1}, y_{j2}, \dots}$, но гистограмма, построенная на ${y_{j1}, y_{j2}, \dots}$ обязанна(с точки зрения нашей модели) иметь нормальное распределение.




# Import section

```python
%matplotlib inline

import matplotlib.pyplot as plt
from collections import defaultdict
from tqdm.notebook import tqdm

from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression, ElasticNet
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

import numpy as np
from scipy.optimize import minimize
```

# CFG

```python
seed = 3
n_samples = 10000
plt.rcParams.update({'font.size': 18})
```

# Helpers

```python
def lr_plot(X,y):
    X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=seed)
    fig, [[ax1, ax2], [ax3, ax4]] = plt.subplots(nrows=2, ncols=2,figsize=(20,10))
    plt.subplots_adjust(hspace = 0.4)
    ax4.axis('off')


    lr = LinearRegression()
    lr.fit(X_train, y_train)
    line_X =  np.arange(X_train.min(), X_train.max(), (X_train.max() - X_train.min())/n_samples)[:, np.newaxis] #np.arange(X_train.min(), X_train.max())[:, np.newaxis]
    line_y = lr.predict(line_X)

    r2 = r2_score(y_test, lr.predict(X_test))
    # regression line and data scatterplot
    ax1.set_title('Fitting curve with data')
    ax1.set(xlabel='X', ylabel='y')
    ax1.plot(line_X, line_y, color='red', linewidth=2, label=f'Linear regressor, $R^2$: {r2:.3f}')
    ax1.scatter(X_train,y_train, label='Train data')

    ax1.set_aspect('auto')
    ax1.grid(True)
    ax1.legend()
    
    # hist of target variable
    ax2.set_title('Histogram of target variable')
    ax2.set(xlabel='y', ylabel='n')
    ax2.hist(y_train, bins=40)
    ax2.set_aspect('auto')
    ax2.grid(True)

    # hist of residuals
    ax3.set_title('Histogram of residuals')
    ax3.set(xlabel='y-y_pred', ylabel='n')
    residuals = y_train-lr.predict(X_train)
    ax3.hist(residuals, bins=40)
    ax3.set_aspect('auto')
    ax3.grid(True)

    plt.show()
  
def lr_plot_polynomial(X,y,order=5):
    X = X.reshape(-1,1)
    lr = LinearRegression()
    X_ = np.column_stack([X**i for i in range(1,order+1)])
    lr.fit(X_ , y)
    plt.scatter(X,y, label='data')
    plt.plot(X, lr.predict(X_ ),color='red', label='regression line')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    plt.legend()
    print(f'Коэффициенты регрессии: {lr.coef_}')
```

# Maximum Likelihood Estimation

```python
X, y, true_coef = make_regression(
    n_samples=n_samples, 
    n_features=1, 
    n_informative=1, 
    noise=3, 
    coef=True, 
    random_state=seed
)

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=seed)

lr = LinearRegression()
lr.fit(X_train, y_train)

def log_likelihood(w):
    return np.sum((y_train-w*X_train.reshape(1,-1))**2)

res = minimize(fun=log_likelihood, x0=0.001, method='BFGS', )


print(f'''
    Настоящий коэффициент: {true_coef:.3f}, r2_score: {r2_score(y_test, true_coef*X_test):.3f}
    Коэффициент полученный методом максимального правдоподобия: {res.x[0]:.3f}, r2_score: {r2_score(y_test, res.x[0]*X_test):.3f}
    Коэффициент полученный регрессией из sklearn: {lr.coef_[0]:.3f}, r2_score: {r2_score(y_test, lr.coef_[0]*X_test):.3f}
    '''
)
```

Данные генерились с помощью sklearn-функции make_regression. Видим, что коэффициент, полученный методом максимального правдоподобия совпадает(либо незначительно отклоняется, если запускать с другим seed, или без него), с коэффициентом из sklearn LinearRegression(), и равен реальному коэф.

Рассмотрим графики: 
* Левый верхний - изображена диаграмма рассеяния обучающей выборки и построена линия линейной регрессии
* Правый верхний - построена гистограмма распределения целевой переменной, просто plt.hist(y)
* Левый нижний - вычислили разность между целевой переменной и прогнозом регресии, далее построили гистограмму остатков plt.hist(y-y_pred)

```python
lr_plot(X,y)
```

Оценивая "на глаз" распределение как целевой переменной, так и остатков  видим, что они близко к нормальному. Теперь рассмотрим другие механизмы порождения данных, не только с нормальным шумом и с более сложной зависимостью y от x.

# Различные модели порождения данных

## Нормальное распределение целевой переменной

Пусть x - случайное целое число из отрезка [0, 99], а y = x + N(0,1), где N(0,1) - нормальный шум с нулевым средним и стандартным отклонением равным 1.

```python
X = np.random.randint(0,100,n_samples)
y = X + np.random.normal(0,1,n_samples)
X = X.reshape(-1,1)
lr_plot(X,y)
```

Видим, что у нас очень хорошая метрика качества на тестовых данных $R^2 = 0.999$, при этом целевая переменная более близка к равномерному распределению, чем к нормальному. А вот распределение остатков близко к нормальному.

## Признак с пуассоновским распределением

Если признак x имеет пуссоновское распределение, а y = x + N(2,1), то распределение целевой переменной близко к нормальному, хотя и немножко смещенно, как и распределение остатков.

```python
X = np.random.poisson(10, n_samples)
y = X + np.random.normal(2,1, n_samples)
X = X.reshape(-1,1)
lr_plot(X,y)
```

## Целевая переменная с пуассоновским шумом

В следующем примере вместо гауссовского шума, мы добавим пуссоновский, при этом гистограмма остатков имеет смещенное распределение с длинным правым хвостом, но метрика качества высока. Можно объяснить это тем, что хвост имеет низкую вероятность и смещение не существенно для модели. 

```python
X = np.random.randint(0,100,n_samples)
y = X + np.random.poisson(2, n_samples)
X = X.reshape(-1,1)
lr_plot(X,y)
```

## Целевая переменная зависит от признака по экспоненциальному закону

Пришло время существенно подпортить данные, пусть теперь y = exp(x+N(0,1)). Диаграмма остатков сильно смещенна, метрика качества упала вниз.

```python
X = np.random.randint(0,10,n_samples)
y = np.exp(X + np.random.normal(0,1,n_samples))
X = X.reshape(-1,1)
lr_plot(X,y)
```

Чтобы исправить ситуацию, можем прологарифмировать y и улучшить метрику:

```python
X = np.random.randint(0,10,n_samples)
y = np.exp(X + np.random.normal(0,1,n_samples))
X = X.reshape(-1,1)

y = np.log(y)
lr_plot(X,y)
```

Т.е. мы просто применили обратную функцию, потому что знали, как генерится y. Если мы этого не знаем, то можем посмотреть на график - экспонента имеет характерный вид.

## Признак имеет логнормальное распределение

Целевая переменная имеет вид y=x+N(0,10). В этом примере видим хорошее качество, но при этом распределение целевой переменной сильно смещенно влево.

```python
X = np.random.lognormal(5,1,n_samples)
y = X + np.random.normal(0,10,n_samples)
X = X.reshape(-1,1)
lr_plot(X,y)
```

В следующих примерах не буду сильно комментировать, мы посмотрим на графики до log-преобразования и посмотрим после, а так же увидим, что преобразование дает прирост в метрик. Код по генерации данных вы можете посмотреть в ноутбуке.

## Целевая переменная распределена нормально, но с обратным средним

### Без преобразования

## $y=(X+1)^5, y=\sqrt{x+1}, y=\frac{1}{x+1}, y=log(1+x), y=log(\frac{1}{X+1})$

Рассмотрим такую зависимость $y=(X+1)^5$, без шума. Если мы добавим еще столбцы со степенями признака $X^2, X^3, X^4, X^5$ и передадим эти данные на вход линейной регрессии, то получим такие коэффициенты:

[ 5. 10. 10.  5.  1.]

Это не что иное, как коэффициенты соответствующего [ряда Тейлора](https://www.wolframalpha.com/input/?i=%28x%2B1%29%5E5+taylor+series)(в данном случае можно просто расскрыть скобки по биному и получить тоже самое)

```python
X =np.arange(0,1,step=1/100).reshape(-1,1)
y = (X+1)**5

lr_plot_polynomial(X,y)
```

Аналогично для [$y=\sqrt{x+1}$](https://www.wolframalpha.com/input/?i=sqrt%28x%2B1%29+taylor+series), при этом добавим чуть больше степеней, чтобы получить лучшую точность на первых пяти коэффициентах

```python
X =np.arange(0,1,0.01)
y = np.sqrt(X+1)

lr_plot_polynomial(X,y, order=8)
```

```python
X =np.arange(0,1,0.01)
y = 1/(1+X)

lr_plot_polynomial(X,y, order=8)
```

Для [$log(x+1)$](https://www.wolframalpha.com/input/?i=log%281%2Bx%29+taylor+series) так же имеет совпадение(с небольшой погрешностью) первых нескольких коэффициентов регрессии с коэф. ряда Тейлора этой функции.

```python
X = np.arange(0,1,0.01)
y = np.log(X+1)
lr_plot_polynomial(X,y, order=8)
```

## $y = (X+1)^5$ c добавлением шума

Но даже небольшое добавление шума сильно влияет на коэффициенты. Возьмем $y=(x+1)^5$ и добавим нормальный шум. Посмотрим, как меняется среднеквадратичная ошибка определения коэффициентов(берем сумму квадратов разностей между реальными коэф. Тейлора и полученными коэф. линейной регрессии, а далее усредняем) в зависимости от параметра распределения $\sigma$. Возьмем выборку в 100 тыс. точек.

```python
sigma_list = []
error_list = []
lr_coef = defaultdict(list)
taylor_coef = [5,10,10,5,1]
X =np.arange(0,1,step=1/100000).reshape(-1,1)
for sigma in tqdm(np.arange(0.0, 1.0, 0.01)):
   
    y = (X+1)**5 + np.random.normal(0, sigma, size=100000).reshape(-1,1)
    X = X.reshape(-1,1)
    lr = LinearRegression(n_jobs=4)
    X_ = np.column_stack([X**i for i in range(1,6)])
    lr.fit(X_ , y)

    for i in range(len(taylor_coef)):
        lr_coef[i+1].append(lr.coef_[0][i])
    
    sigma_list.append(sigma)
    error_list.append(0.2*np.sum((lr.coef_-np.array(taylor_coef))**2))
    
plt.plot(sigma_list, error_list)
plt.xlabel('$\sigma$')
plt.ylabel('MSE')
plt.grid()
plt.show()
```

Видно, что даже в небольшем диапазоне изменения $\sigma$ - от 0 до 1.0, среднеквадратичная ошибка существенно возрастает. При этом коэффициенты регрессии все больше отклоняются от своих истинных значений:

```python
for i in lr_coef:
    plt.plot(sigma_list, lr_coef[i])
    plt.axhline(y=taylor_coef[i-1], color='r', linestyle='-')
    plt.xlabel('$\sigma$')
    plt.ylabel('regression coef.')
    plt.title(f'True coef: {taylor_coef[i-1]}')
    plt.grid()
    plt.show()
```
