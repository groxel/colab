## Лекция 1

- ханипот - добавляем в дотаяет для разметки заранее нами размеченные данные, чтобы проверить качество работы разведчиков
- тестовая выборка должна быть репрезентативной? Достаточно большой/ подбирается индивидуально под задачу
- кросс-валидация
 - полезно, когда выборка небольшая (несколько тыс)
 - разбиваем дотаяет на части (например на 3 - это компромисс между скоростью и качеством результата)
 - затем делаем обучение на 2х? Тес на одном - получаем качество Q1
 - меняем части местами - получаем Q2, Q3
 - потом все Q  усредняются или берем лучшую и обучаем на всех даннызх

- данные для тестовой выборки должны со временем меняться (если долго подбирать алгоритмы под одну и ту же тестовую выборку). Чтобы модель не подгоняла результаты под тесты

Классы моделей

- у функции ошибки может быть несколько локальных минимумов
 - все зависит от точки старта
 - найти глобальный минимум - нерешенная задача, но в случае хорошей выборки данных минимум будет один
 - поиск минимуму - градиентный спуск
 - аналитическое решени будет более трудоемким при большом кол-ве параметров (тысячи, миллионы), чем градиентный спуск

- добавление нелинейных признаков часто улучшая модель - но это ухудшая интерпретируемость модели
 - кол-во периметров в модели увеличивает время обучение скорее линейно
 - подбор нелинейных параметров в регрессии модели - это просто творческий поиск

Метрики качества моделей

- через функцию ошибки мы задаем требования к модели

- MSE
 - проблема в единицах измерений, поэтому берут корень из mse (RMSE)
 - стремится улучшить модель на выбросах, но ухудшает на нормальных данных - вместо MSE лучше использовать MSA

- коэффициент R^2
 - позволяет оценить становится ли наша модель лучше или хуже

- MSA
 - при наличии выбросов лучше использовать  MSA вместо MSE

При прогнозировании лучше применять относителньую ошибку (MAPE).

Можно придумать свою метрику, которая будет отражать какие-то экономически енюансы бизнеса.

**Классификация**

В регрессии измеряем ошибку, а в классификации качество.

Задать вопрос: какое соотношение классов в выборке. Чем хуже соотношение, тем хуже выборка и меодль

Матрица ошибок строится по новой модели на исторических данных (анпример выдача кредитов)

Точность (precsicion) - отношение возвращенных кредитов ко всем выданным

Полнота (recall) - отношение возвращенных кредитов к возвращенным+потенциал

Бизнес метрика **LTV** (lifetime value)
up-sell - более дорогие товары
cross-sell - сопутствующие товары

ЛТВ долго считается. А вот конверсия быстро. Нам нужны онлайн метрики.
Офлайн метрики - считаем по историческим данным.

**Ухудшающие эксперименты**
- используем когда не понятно какую таргет метрику использовать
- ухудшаем продукт для части пользователей
- подбираем модуль так. которая будет гарантированно ухудшаться при ухудшении продукта
- в таком случае такие модули часто хорошо работают и при улучшении продукта

Решающий пень - дерево глубиной 1

Гр бустинг в отличии от обычного случайного леса может переобучиться, т е станет хуже при увеличении кол-ва деревьев.

**Кластеризация**

K-means
- хорошо находит только кластеры круглой формы
- плохо работает с кластерами сложной формы
- результаты сильно зависят от первоначального расположения K точек (которые определяются случайно)
- тут тоже есть функция потерь - парабола
- метод быстрый
- есть гарантия, что алгоритм конечный (благодаря функции потерь)

DBSCAN
- работает медленно
- все равно откуда начинать поиск, результат будет одинаковый
- результат зависит от гипермараметров
- работает с кластерами любой формы
- алгоритм сам определяем кол-во кластеров - плохо интерпретировать результаты


**Визуализация / снижение размерности РСА**

Снижение размерности
- помогает избавиться от признаков-шумов
- помогает визуализировать данные (приведя их к двух или трех мерному пространству)
- метод РСА (метод главных компонент) можно использовать для анонимизации данных
- главные компоненты - это некая линейная комбинация исходных признаков, потом через них можно восстановить исходные данные?

Визуализация t-SNE
- нелинейный, вероятностный подход
- лучше чем РСА
- но работает существенно дольше

Визуализация очень полезна для интерпретации тех кластеров, что получились в результате кластеризации.
